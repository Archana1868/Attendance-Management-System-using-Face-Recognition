```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
face_cascade = cv2.CascadeClassifier('Haarcascades\haarcascade_frontalcatface.xml')
```

```python
imgpath = 'Images\sample5.jpg'
```

```python
img = cv2.imread(imgpath)
```

```python
img=cv2.resize(img,(800,600))
```

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
faces = face_cascade.detectMultiScale(gray_img,scaleFactor = 1.01, minNeighbors=7)
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(gray_img,(x,y),(x+w,y+h),(255,255,255),3)
```

```python
plt.figure(figsize=(12,8))
plt.imshow(gray_img,cmap='gray')
plt.show()
```
